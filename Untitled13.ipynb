```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/theresathankachan/S3MCA/main/Datascience/Datasets/iris.csv")
print(df.head(10))
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(subset="Species")
```

```python
df.value_counts("Species")
```

```python
df=df.drop(columns=['Id'])
```

```python
sns.countplot(x='Species', data=df, )
plt.show()
```

```python
#scatterplot
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',
                hue='Species', data=df, )
```

```python
#scatterplot
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',
                hue='Species', data=df, )
```

```python
#bubble chart
x = df["SepalLengthCm"].to_numpy()
y = df["PetalLengthCm"].to_numpy()
z = df["SepalLengthCm"].to_numpy()
colors = np.random.rand(len(z)) 

plt.scatter(x, y,c=colors)


plt.show()
```

```python
#scatterplot
sns.scatterplot(x='Species', y='PetalWidthCm',
                hue='Species', data=df, )
```

```python
#Histogram
df['SepalLengthCm'].hist()
```

```python
#Histogram
df['SepalWidthCm'].hist()
```

```python
#Histogram
df['PetalLengthCm'].hist()
```

```python
#Histogram
df['PetalWidthCm'].hist()
```

```python
#Andrews curve
x=pd.plotting.andrews_curves(df,'Species')
x.plot()
plt.show()
```

```python

# Density Chart

#df["PetalLengthCm"].plot(kind="density")
df.plot(kind='density')
plt.show()
```

```python
#Parallel plotting
pd.plotting.parallel_coordinates(df, "Species")
plt.ioff()
```

```python
#distribution chart
sns.displot(df, x = "Species", bins= 10)
plt.title("Species counts")
plt.show()
```

```python
ax = sns.lineplot(x="PetalLengthCm",y="PetalWidthCm",data=df,hue="Species")
```

```python
#box plot
sns.boxplot(x="Species",y="PetalLengthCm",data=df)
plt.show()
```

```python
df.boxplot(by="Species",figsize=(12,10))
```

```python
#scatter matrix
sns.pairplot(df,hue="Species",height=3)
```

```python
#scatter multiple
ax = df.plot(kind="scatter", x="SepalLengthCm",y="SepalWidthCm", color="b", label="SepalLengthCm vs SepalWidthCm")
df.plot(kind="scatter", x="SepalLengthCm",y="PetalWidthCm", color="r", label="SepalLengthCm vs PetalWidthCm", ax=ax)
df.plot(kind="scatter", x="SepalLengthCm",y="PetalLengthCm", color="g", label="SepalLengthCm vs PetalLengthCm", ax=ax)

ax.set_xlabel("SepalLengthCm")
ax.set_ylabel("vertical label")
plt.show()
```
